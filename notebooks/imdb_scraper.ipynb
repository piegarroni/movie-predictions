```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
# loading datasets and preliminary cleaning

metas=pd.read_csv('../../data/movies_meta.csv')
metas=metas.loc[metas['original_language']=='en']  # only english movies
metas=metas.drop(['production_countries', 'overview', 'tagline', 'belongs_to_collection', 'homepage', 'revenue', 'spoken_languages', 'video', 'homepage', "poster_path", 'production_companies'], axis=1)
print(metas.__len__())
```

```python
ids=list(metas['imdb_id'])
print(ids[:5])
```

```python

from multiprocessing import Pool
import requests
from bs4 import BeautifulSoup

base_url = 'http://www.imdb.com/title/'

all_urls = []
votes={}

def generate_urls():
    for i in ids:
        all_urls.append(base_url + str(i))
    
def scrape(url):    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    a = soup.find("span", {"class": "sc-7ab21ed2-1 jGRxWM"})
    print(a, url.split("/")[-2])
    votes[url.split("/")[-1]]=a  
   

generate_urls()

p = Pool(10)
p.map(scrape, all_urls)
p.terminate()
p.join()
```

```python


urls=[]
votes=[]

# drop nan ids!!!!!!!!!!
for i in metas['imdb_id'].dropna():
    #print(i) 
    urls.append('http://www.imdb.com/title/'+i)
    response = requests.get('http://www.imdb.com/title/'+i)
    soup = BeautifulSoup(response.text, "html.parser")
    a = soup.find("span", {"class": "sc-7ab21ed2-1 jGRxWM"})
    votes.append(a)  
 
```

```python
print(len(metas), len(votes))
```

```python
metas['rating']=pd.Series(votes)
```
